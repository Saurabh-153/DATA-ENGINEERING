## Slicing ==> group of elements which are in order

### 1 D Slicing

**Python's slice operator:**<br>

Syntax-1: l[begin:end]<br>
• It returns elements from begin index to (end-1) index

```python
Image(filename=r"C:/Users/saurabhkumar9/numpy_image/slicing.JPG")
```

```python
l = [10,20,30,40,50,60,70]

print(l[2:6],'\n') #from 2nd index to 5th index
print(l[-6:-2], '\n') #From -6 index to -3 index

# If we are not specifying begin index then default is 0
print(l[:3], '\n') #from 0 index to 2nd index

# If we are not specifying end index then default is upto last
print(l[2:], '\n') # from 2nd index to last

# if we are not specifying begin and end then it will considered all elements
print(l[:]) # it is equivalent to l
```

Syntax-2: l[begin:end:step]<br>
• begin,end and step values can be either +ve or -ve. These are optional<br>
• If begin is not specified then 0 will be taken by default<br>
• If end is not specified then upto end of the list will be taken<br>
• If step is not specified then 1 will be taken by default<br>
• step value cannot be 0<br>
• If step value is +ve, then in forward direction begin to end-1 will be considered<br>
• If step value is -ve, then in backward direction begin to end+1 will be considered<br>
• In forward direction, if end is 0 then the result will be always empty<br>
• In backward direction, if end is -1 then the result will be always empty<br>

Note:<br>
• slice operator does not raise any IndexError when the index in out of range<br>

```python
# slice operator in python list
l = [10,20,30,40,50,60,70,80,90]
print(l[2:6:2], '\n') # from 2nd index to 6-1 index with step 2

# Slice operator on 1-D array
import numpy as np
a = np.arange(10,101,10)
print(a,'\n')

print(a[2:5],'\n') # from 2nd index to 5-1 index
print(a[::1],'\n') # entire array
print(a[::-1]) # entire array in reverse order
```

## Advanced indexing
group of elements which are not in order (arbitrary elements)

To access arbitrary elements(elements which are out of order) we should go for Advanced Indexing.<br>

By using **index**, we can access only one element at a time.<br>
• 1-D array ==> a[i]<br>
• 2-D array ==> a[i][j]<br>
• 3-D array ==> a[i][j][k]<br>

By using **slice operator** we can access multiple elements at a time, but all elements should be in order.<br>
• 1-D array ==> a[begin:end:step]<br>
• 2-D array ==> a[begin:end:step,begin:end:step]<br>
• 3-D array ==> a[begin:end:step,begin:end:step,begin:end:step]<br>

### 1 D Array

**Syntax** :<br>
• array[x] ==> x can be either ndarray or list, which represents required indices <br>

**1st Way(ndarray)**<br>
• create an ndarray with required indices in the given original array<br>
• pass that array as argument to the original array<br>

**2nd way(list)**<br>
• create a list with the required indices in the given original array<br>
• pass that list as argument to the original array

```python
a = np.arange(10,101,10)
print(a, '\n')

# 1st way (ndarray)

# step1 : create an ndarray with the required indices
# assume that we have to extract 30,50,60,80 elements. Their indices are 2,4,5,8
# create an ndarray with those indices
indices = np.array([2,4,5,8])
print(indices, '\n')

# step 2: pass the indices as argument to the original array to get the required arbitrary elements
print(a[indices])
```

```python
# 2nd way (list)

# Step 1: create a list with the required indices of the original array
l = [2,4,5,8]

#Step 2: pass the list as an argument to the original array to get the required arbitrary elements
print(a[l], '\n')
```

## Condition based selection : array_name[condition]

We can select elements of an array based on condition also

### Syntax :: array_name[boolean_array]

In boolean_array, whereever True is present the corresponding value will be selected

```python
import numpy as np
a = np.array([10,20,30,40])
boolean_array=np.array([True,False,False,True])
a[boolean_array]
```

### Syntax for condition based selection :: array_name[condition]

Condition always return boolean value

```python
# select elements which are greater than 25 from the given array
a = np.array([10,20,30,40]) # boolean_array should be [False,False,True,True]
# condition will give the boolean value
print(a>25, '\n')

# create boolean array
bool_array = a>25
# pass the bool_array as argument to the original array
print(a[bool_array], '\n')

# in single step also we can achieve the functionality
# array_name[condition]
print(a[a>25], '\n')

# select then negative numbers from the given array
a = np.array([10,-5,20,40,-3,-1,75])
print(a[a<0], '\n')

## condition based selection is applicatble for 2-D array also
a = np.arange(1,26).reshape(5,5)
print(a, '\n')

# select the even numbers from 2-D array
print(a[a%2==0])
```

### Logical operations

• **logical_and(x1, x2, /[, out, where, …]) ==>** Compute the truth value of x1 AND x2 element-wise.<br>
• **logical_or(x1, x2, /[, out, where, casting, …])** ==> Compute the truth value of x1 OR x2 element-wise.<br>
• **logical_not(x, /[, out, where, casting, …])** ==> Compute the truth value of NOT x element-wise.<br>
• **logical_xor(x1, x2, /[, out, where, …])** ==> Compute the truth value of x1 XOR x2, element-wise.

```python
help(np.logical_and)
```

```python
# we can select elements by using multiple conditions by using logical operations
# select the elements which are even and divisible by 5
a = np.arange(1,26).reshape(5,5)
b = np.logical_and(a%2==0,a%5==0)
print(f"Original array a ==> \n {a}")
print(f"Logical array b ==> \n {b}")
print(f"Elements which are even and divisible by 5 ==> \n {a[b]}")
```

### Selecting elements based on multiple conditions

We can use & for AND condition and | for OR condition<br>
• array_name[(condition_1) & (condition_2)]<br>
• array_name[(condition_1) | (condition_2)]

```python
# select the elements which are divisible by 2 and divisible by 3
a = np.arange(1,26).reshape(5,5)
print(a, '\n')
print(a[(a%2 == 0) & (a%3 ==0)], '\n')

bool_array = np.logical_and(a%2==0,a%5==0)
print(a[bool_array], '\n')
```

```python
# select the elements either even and divisible by 3
a = np.arange(1,26).reshape(5,5)
print(a)
a[np.logical_or(a%2==0,a%3==0)]
```

```python
# select the elments which are divisible by 2 and 3
a = np.arange(1,26).reshape(5,5)
a[(a%2 == 0) & (a%3 ==0)]
```

```python
# select the elments which are divisible by either 2 or 3
a = np.arange(1,26).reshape(5,5)
a[(a%2 == 0) | (a%3 ==0)]
```

### Python Slicing Vs Numpy Array Slicing Vs Adv. Indexing Vs. Conditional Selection

#### case-1 : Python slicing

• In the case of list, slice operator will create a separate copy.<br>
• If we perform any changes in one copy those changes won't be reflected in other copy.

```python
# Python Slicing ==> new copy is created
l1 = [10,20,30,40]
l2=l1[::]
l1 is l2
```

```python
# if we made changes in l1 it does not reflect on l2 and vice-versa
l1[0] = 888
print(f"l1::{l1}")
print(f"l2::{l2}")
```

```python
l2[1] = 7777
print(f"l1::{l1}")
print(f"l2::{l2}")
```

#### case-2 : Numpy Array Slicing

• A separate copy won't be created and just we are getting view of the original copy.<br>
• View is logical entity where as Table is physical entity.(RDBMS)<br>

```python
# Numpy array slicing
a = np.arange(10,101,10)
a
```

```python
# slicing on numpy array
b = a[0:4] # view is created
print(f"b ==> {b}")
print(f"a ==> {a}")
```

```python
# if we made changes in b it will reflect on a also vice-versa
b[0] = 999
print(f"After modifying the b array value ==> {b}")
print(f"a array value ==> {a}")
```

```python
a[2] = 888
print(f"After modifying the a array value ==> {a}")
print(f"b array value ==> {b}")
```

#### case-3 : Advanced Indexing and Condition Based Selection
    
• It will select required elements based on provided index or condition and with those elements a new 1-D array object will be created.<br>
• The output is always a new 1-D array only.

```python
# Advanced Indexing
a = np.arange(10,101,10)
b = a[[0,2,5]]
print(f" a ==> {a}")
print(f" b ==> {b}")
```

#### Slicing Vs Advanced Indexing

| Feature             | Slicing                                     | Advanced Indexing                          |
|---------------------|---------------------------------------------|--------------------------------------------|
| Element Order       | The elements should be ordered              | The elements need not be ordered           |
| Element Selection   | We can't select arbitrary elements          | We can select arbitrary elements           |
| Selection Type      | Conditional based selection is not possible | Conditional based selection is possible    |
| Nature of Output    | Just we will get view but not copy          | Just we will get separate copy but not view|
| Memory Performance  | Memory, performance-wise it is the best     | Memory, performance-wise it is not up to the mark |

#### Summary of syntaxes

| Indexing Type          | 1-D Array                                | 2-D Array                                       | 3-D Array                                                   |
|------------------------|------------------------------------------|------------------------------------------------|------------------------------------------------------------|
| **Basic Indexing**     | `a[i]`                                   | `a[i][j]` or `a[i,j]`                          | `a[i][j][k]` or `a[i,j,k]`                                  |
| **Slicing**            | `a[begin:end:step]`                      | `a[begin:end:step, begin:end:step]`            | `a[begin:end:step, begin:end:step, begin:end:step]`         |
| **Advanced Indexing**  | `a[x]` (x contains required indices)     | `a[[row indices], [column indices]]`           | `a[[indices of 2D array], [row indices], [column indices]]` |
| **Condition based**    | `a[condition]` (e.g., `a[a>0]`)          | Same as 1-D array                               | Same as 1-D array                                           |

# Chapter-6 - How to iterate elements of nd array

#### Iterate Elements of nd Array

- **Iteration Definition**: Iteration means getting all elements one by one.

#### We can iterate nd arrays in 3 ways:

1. **By using Python's loops concept**
2. **By using nditer() function**
3. **By using ndenumerate() function**

## By using Python's loops concept

```python
# Iterate elemetns of 1-D array

import numpy as np
a = np.arange(10,51,10)
for x in a:
    print(x)
```

```python
# Iterate elemetns of 2-D array

import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(a, '\n')
for x in a: #x is 1-D array but not scalar value
    for y in x: # y is scalar value present in 1-D array
        print(y)
```

```python
# Iterate elemetns of 3-D array

import numpy as np
a = np.array([[[10,20],[30,40]],[[50,60],[70,80]]])
print(a,'\n')
for x in a: #x is 2-D array but not scalar value
    for y in x: # y is 1-D array but not scalar value
        for z in y: # z is scalar value
            print(z)
```

Note: To iterate elements of n-D array, we require n loops.

## By using nditer() function

**Advantage**: To iterate any n-D array only one loop is enough.<br>

• nditer is a class present in numpy library.<br>
• nditer() ==> Creating an object of nditer class.

```python
# Iterate elements of 1-D array

import numpy as np
a = np.arange(10,51,10)
for x in np.nditer(a):
    print(x)
```

```python
# Iterate elements of 2-D array

import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(a,'\n')
for x in np.nditer(a):
    print(x)
```

```python
# Iterate elements of 3-D array

import numpy as np
a = np.array([[[10,20],[30,40]],[[50,60],[70,80]]])
print(a,'\n')
for x in np.nditer(a):
    print(x)
```

### Iterate elements of sliced array also

```python
import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(a,'\n')
for x in np.nditer(a[:,:2]):
    print(x)
```

### Using nditer() to get elements of required data 

• We have to use op_dtypes parameter

```python
import numpy as np
help(np.nditer)
```

```python
import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
for x in np.nditer(a,op_dtypes=['float']):
    print(x)
print(a)
```

```python
# Numpy won't change the type of elements in existing array.
# To store changed type elements, we required temporary storage, which is nothing but buffer.
# We have to enable that buffer.
import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
for x in np.nditer(a,flags=['buffered'],op_dtypes=['float']):
    print(x)
print('\n', a)
```

```python
import numpy as np
a = np.array([[[10,20],[30,40]],[[40,50],[60,70]]])
for x in np.nditer(a):
    print(x.dtype) #int32
```

```python
for x in np.nditer(a,flags=['buffered'],op_dtypes=['int64']):
    print(x.dtype) #int64
```

### Normal Python's loops vs nditer()

**Python Loops**<br>

**nditer()**<br>

## By using ndenumerate() function

• By using **nditer()** we will get **elements** only but **not indexes**<br>
• If we **want indexes also** in addition to elements, then we should use **ndenumerate()** function.<br>
• **ndenumerate()** function returns **multidimensional index iterator** which yields pairs of array indexes(coordinates) and values.

```python
# Iterate elements of 1-D array:
import numpy as np
a = np.array([10,20,30,40,50,60,70])
for pos,element in np.ndenumerate(a):
    print(f'{element} element present at index/position:{pos}')
```

```python
# Iterate elements of 2-D array:

import numpy as np
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
for pos,element in np.ndenumerate(a):
    print(f'{element} element present at index/position:{pos}')
```

```python
# Iterate elements of 3-D array:

import numpy as np
a = np.arange(1,25).reshape(2,3,4)
for pos,element in np.ndenumerate(a):
    print(f'{element} element present at index/position:{pos}')
```
