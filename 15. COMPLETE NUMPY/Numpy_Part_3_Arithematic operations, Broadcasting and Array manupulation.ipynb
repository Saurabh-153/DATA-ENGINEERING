# Chapter-8 - Broadcasting

**Broadcasting**<br>

• Generally Arithmetic operations are performed between two arrays are having same dimension, shape and size.<br>
• Eventhough dimensions are different,shapes are different and sizes are different still some arithmetic operations are allowed by Broadcasting<br>
• Broadcasting will be performed automatically by numpy itself and we are not required to perform explicitly.<br>
• Broadcasting won't be possible in all cases.<br>
• Numpy follow some rules to perform Broadcasting. If the rules are satisfied then only broadcasting will be performed internally while performing arithmetic operations.<br>

**Note**<br>
• If both arrays have same dimension, same shape and same size then broadcasting is not required.<br>
• Different dimensions or different shapes or different sizes then only broadcasting is required.<br>

| Broadcasting Rule | Description                                                                                                         | Example Before             | Example After                                          |
|-------------------|---------------------------------------------------------------------------------------------------------------------|----------------------------|--------------------------------------------------------|
| **Rule-1**        | Ensure both arrays have the same dimension. If not, pad the shape of the lesser dimension array with 1's on the left. | `a` shape: (4,3), `b` shape: (3,) | `a` shape: (4,3), `b` shape: (1,3)                   |
| **Rule-2**        | If the sizes of the two arrays do not match in any dimension, expand the smaller size to match the larger one.        | `a` shape: (4,3), `b` shape: (1,3) | `a` shape: (4,3), `b` shape: (4,3)                   |

**Note:**
- Arrays with a size not equal to 1 in any unmatched dimension will trigger an error, as Numpy cannot perform broadcasting.
- Data is reused from the input array for the necessary dimension (rows or columns).
- The result will always have the higher dimension of the input arrays, allowing for compatible arithmetic operations.

**Eg: Broadcasting between (3,2,2) and (3,) possible or not**<br>

| Stage                   | Array `a` Details    | Array `b` Details    | Action Taken by Numpy                                        | Result                                   |
|-------------------------|----------------------|----------------------|--------------------------------------------------------------|------------------------------------------|
| **Before Rule-1**       | Shape: (3,2,2)       | Shape: (3,)          | Different dimensions.                                        |                                          |
| **After Rule-1 Applied**| Shape: (3,2,2)       | Shape becomes: (1,1,3) | Align dimensions by padding `b` with 1's on the left.       | Both arrays now have the same dimension. |
| **Before Rule-2**       | Shape: (3,2,2)       | Shape: (1,1,3)       | Arrays are in the same dimension but different sizes.        |                                          |
| **After Rule-2 Applied**| Shape: (3,2,2)       | Shape becomes: (3,2,3) | Expand `b` where size 1 to match corresponding `a` sizes.   | Both arrays now have the same dimensions but different shapes. |
| **Final Result**        | Shape: (3,2,2)       | Shape: (3,2,3)       | Broadcasting not possible due to different final shapes.     | Numpy unable to perform broadcasting.    |

**Note:**
- Numpy requires final shapes to be exactly the same for broadcasting to successfully occur.
- In this example, despite both arrays being adjusted to the same dimension, their final shapes differ, thus broadcasting is not achievable.

```python
# Broadcasting between 1-D arrays of different sizes
import numpy as np
a = np.array([10,20,30,40])
b = np.array([1,2,3])
# a : (4,) b: (3,).
# Both are having same diemensions. But sizes are different so arithmetic operation is not performed
# Broadcasting by Numpy will be failed while performing arithmetic operation in this case
a+b
```

## reshape() function
• We can use reshape() function to change array shape without changing data.

| Feature                 | Details                                                        |
|-------------------------|----------------------------------------------------------------|
| **Function**            | `reshape(a, newshape, order='C')`                              |
| **Purpose**             | Gives a new shape to an array without changing its data.       |
| **Parameters**          | `newshape`: int or tuple of ints                               |
|                         | `order`: {'C', 'F', 'A'}                                       |
| **Default Value**       | `order='C'`                                                    |
| **Order 'C'**           | C language style, row major order.                             |
| **Order 'F'**           | Fortran language style, column major order.                    |
| **Data Integrity**      | The data should not be changed. Input size and output size should match, otherwise a `ValueError` is raised. |
| **Memory**              | No change in the data, thus no new object is created. The result is a view, not a copy. |
| **Mutability**          | Changes to the original array are reflected in the reshaped array and vice versa. |
| **Unknown Dimension**   | You can specify an unknown dimension size as `-1`, but only once. |
| **Usage**               | `np.reshape(a, new_shape)` - Functional style                  |
|                         | `ndarray_obj.reshape(new_shape)` - Object-oriented style       |

**Note:**
- The `reshape` function can be used either as a standalone function (`np.reshape()`) or as a method of an `ndarray` object (`ndarray_obj.reshape()`).
- Both usage styles will consider the specified order for reshaping: 'C' for row-major and 'F' for column-major.

| Feature              | Details                                                     |
|----------------------|-------------------------------------------------------------|
| **Functionality**    | Reshaping arrays from one shape to any other valid shape    |
| **Examples**         | (10,) to (5,2), (2,5), (10,1), (1,10); (24,) to (3,8), (6,4), (2,3,4), (2,2,2,4) |
| **Implementation**   | Available both as a method in the ndarray class and as a standalone function in the numpy module. |
| **Numpy Module**     | `numpy` - The module where ndarray and reshape are located. |
| **ndarray Class**    | Contains the `reshape()` method, can be called on any ndarray object. |
| **reshape() Method** | Called on ndarray objects: `b = a.reshape(shape, order)`    |
| **reshape() Function** | Standalone function in numpy: `b = np.reshape(a, shape, order)` |

**Note:**
- Reshaping an array allows for multiple views of the same data, facilitating diverse data manipulation without copying the data.
- The ability to reshape arrays is fundamental in numpy for efficient data handling and transformation.

### by using np.reshape() ==> Functional Style

```python
# While using reshape() function, make sure the sizes should be matched otherwise we will get error.

import numpy as np
a = np.arange(1,11)
print(f"array : {a}")
b = np.reshape(a,(5,3))
```

```python
# converting 1-D array to 2-D array

a = np.arange(1,11)
print(f"array : {a}")
b = np.reshape(a,(5,2))
print(f"Converting 1-D(a) array to 2-D(b) array : \n {b} ")
```

```python
# View is created but not copy Any change in the original array will be reflected to the reshaped
# array

a = np.arange(1,16)
b = np.reshape(a,(5,3))
print("*"*80)
print(f"Original array : {a}")
print(f"Reshaped array :\n {b}")
print("*"*80)
print("After modification in the orginal array")
a[0] = 111
print(f"Original array : {a}")
print(f"Reshaped array :\n {b}")
print("*"*80)
print("After modification in the reshaped array")
b[2,1] = 888
print(f"Reshaped array :\n {b}")
print(f"Original array : {a}")
print("*"*80)
```

### by using ndarray_obj.reshape() ==> Object oriented style

```python
a = np.arange(1,25)
b = a.reshape((2,3,4))
print("*"*80)
print(f"Original array : \n {a}")
print(f"Reshaped array :\n {b}")
print("*"*80)
```

```python
# C style order :: Row major order
# It will consider 1st row, 2nd row and so on

a = np.arange(12)
b = a.reshape((3,4))
print("*"*80)
print(f"Original array : {a}")
print(f"Reshaped array :\n {b}")
print("*"*80)
```

```python
# Fortran style order :: Column major order
# It will consider 1st column, 2nd column and so on

a = np.arange(12)
b = a.reshape((3,4),order='F')
print("*"*80)
print(f"Original array : {a}")
print(f"Reshaped array :\n {b}")
print("*"*80)
```

### Unknown Dimension with `-1` in Numpy's Reshape Function

- **Single Usage of `-1`**: You can use `-1` for one dimension only. This allows numpy to automatically calculate the necessary size for that dimension based on the array's total size and the other specified dimensions.
- **Automatic Calculation**: Numpy decides and replaces `-1` with the required dimension. This feature is particularly useful when you know the total number of elements in the array but want flexibility in how dimensions are defined.
- **Constraint on Usage**: Only one shape dimension can be `-1`. If `-1` is used more than once, it results in an invalid operation.
- **Valid Examples**:
  - `b = a.reshape((5,-1))`: Numpy calculates the second dimension automatically.
  - `b = a.reshape((-1,5))`: Numpy calculates the first dimension automatically.
- **Invalid Example**:
  - `b = a.reshape((-1,-1))`: Using `-1` more than once is invalid and will raise an error.

**Note**: The ability to use `-1` in the reshape function adds significant flexibility to array manipulation, allowing for dynamic adjustment of array shapes without detailed manual calculation of each dimension.

```python
a = np.arange(1,11)
b = a.reshape((5,-1))
print(f"Shape of array b :: {b.shape}")
```

```python
a = np.arange(1,11)
b = a.reshape((-1,5))
print(f"Shape of array b :: {b.shape}")
```

```python
a = np.arange(1,11)
b = a.reshape((-1,-1))
```

| Feature                  | `numpy.resize()`                               | `ndarray.resize()`                             |
|--------------------------|----------------------------------------------|----------------------------------------------|
| **Functionality**        | Changes the size of an array and fills in with repeated copies of `a` along each axis if necessary. | Alters the shape and size of an array in-place. |
| **Data Handling**        | Creates a new array with the specified shape, and fills it with repeated copies of the input array i.e. if new array requires more element then repeted copy of original array will be used. | Modifies the array itself; does not return a new array, if more element is needed it is filled with zero  |
| **Return Type**          | Create and returns a new array with the specified shape. | It won't return new array and exusting array will be modified (modifies in-place).                    |
| **Memory Allocation**    | Allocates new memory for the new array.      | Does not allocate new memory; tries to use existing memory space. |
| **Usage**                | `numpy.resize(a, new_shape)` - does not modify the original array `a`. | `a.resize(new_shape)` - directly modifies the array `a`. |
| **Behavior with Excess Elements** | If the new size is larger, repeats the original array as needed. | If the new size is larger without enough elements, it fills with zeros. |
| **Behavior with Fewer Elements** | Trims the excess elements to fit the new size. | Reduces the size and drops the excess elements. |
| **Flexibility**          | Can handle resizing to a completely different size not constrained by the number of original elements. | Limited to adjustments that can fit within the original array's allocated memory, unless it's a reduction. |

**Note**:
- `numpy.resize` is more flexible for creating a completely new array configuration and is non-destructive to the original array.
- `ndarray.resize` is useful for modifying an array size in-place, particularly when you are certain about memory constraints and do not need the original array configuration.

## flatten() method

**Syntax**: ndarray.flatten(order='C')

| Feature                 | Details                                                         |
|-------------------------|-----------------------------------------------------------------|
| **Functionality**       | Collapses an n-Dimensional array into a 1-Dimensional array.    |
| **Return Type**         | Returns a new 1-Dimensional array (a copy, not a view).         |
| **Memory Handling**     | Creates a new array, hence changes in the original array do not affect the flattened array and vice versa. |
| **Dimensionality**      | The output of the `flatten` method is always a 1-D array.       |
| **Order Options**       | `C-style`: Row major order, `F-style`: Column major order       |
| **Method Availability** | Available as a method in the `ndarray` class (`a.flatten()`) but not as a standalone function in the numpy library (`np.flatten()` is invalid). |

**Note**:
- `flatten` method is useful for converting a multi-dimensional array into a one-dimensional array for easier processing and manipulation.
- Since `flatten` returns a copy, it is safe to use without altering the original data structure.

```python
import numpy as np
help(np.flatten)
```

```python
help(np.ndarray.flatten)
```

```python
# converting 2-D array to 1-D array ==> C style
a = np.arange(6).reshape(3,2)
b = a.flatten()
print(f"Original array :\n {a}")
print(f"Flatten array :\n {b}")
```

```python
# change the value of Original array. The changes wont be reflected in Flatten array
a[0,0] = 8888
print(f"Original array :\n {a}")
print(f"Flatten array :\n {b}")
```

```python
# converting the 2-D array to 1-D array ==> Fortran style F
a = np.arange(6).reshape(3,2)
b = a.flatten('F')
print(f"Original array :\n {a}")
print(f"Flatten array :\n {b}")
```

```python
# Converting 3-D to 1-D ==> C Style
a = np.arange(1,19).reshape(3,3,2)
b = a.flatten()
print(f"Original array :\n {a}")
print(f"Flatten array :\n {b}")
```

```python
# Converting 3-D to 1-D ==> Fortran Style(F)
a = np.arange(1,19).reshape(3,3,2)
b = a.flatten('F')
print(f"Original array :\n {a}")
print(f"Flatten array :\n {b}")
```

| 1-D array `(4,)`         | `(4,)` No change           |
| 2-D array `(2,3)`        | `(3,2)`                    |
| 3-D array `(2,3,4)`      | `(4,3,2)`                  |
| 4-D array `(2,2,3,4)`    | `(4,3,2,2)`                |

**Note**:
- When calling the `transpose()` function without specifying the 'axes' parameter, the dimensions of the input array are simply reversed.
- This behavior applies consistently across different dimensional arrays, effectively reversing the order of dimensions.
- For 1-D arrays, since there is only one dimension, the transpose operation does not change the array.

```python
# Tranpose of 1-D array ==> no impact
a = np.array([10,20,30,40])
atrans = np.transpose(a)
print(f"Original Array : \n {a}")
print(f"Transposed Array : \n {atrans}")
print(f"Original Array shape : {a.shape}")
print(f"Transposed Array shape: {atrans.shape}")
```

```python
# Transpose of 2-D
# In 2-D array, because of transpose, rows will become columns and columns will become rows.
a = np.arange(1,7).reshape(2,3)
atrans = np.transpose(a)
print(f"Original Array : \n {a}")
print(f"Transposed Array : \n {atrans}")
print(f"Original Array shape : {a.shape}")
print(f"Transposed Array shape: {atrans.shape}")
```

## reshape Vs transpose

• In reshape we can change the size of dimension, but total size should be same.

| Description | Details |
|-------------|---------|
| **Input and Transpose Output Examples** | |
| **Input:** `(3,4)` | **Output:** `(4,3)` |
| **Input:** `(2,3,4)` | **Output:** `(4,3,2)`, `(2,4,3)`, `(3,2,4)`, `(3,4,2)` *(Invalid examples: `(2,12)`, `(3,8)` as these change the size of dimensions)* |
| **Functionality** | Transpose is a specific case of reshape where the dimensions are interchanged but their sizes are not altered. |
| **Syntax Variations for Transpose** | |
| `numpy.transpose(a)` | Transposes the array `a`. |
| `numpy.transpose(a, axes=(2,0,1))` | Transposes the array `a` according to the order specified in `axes`. |
| `ndarrayobject.transpose()` | Transposes the ndarray object. |
| `ndarrayobject.transpose(*axes)` | Transposes the ndarray object according to the order specified in `axes`. |
| `ndarrayobject.T` | A shorthand for transposing an ndarray object. |
| **Note** | Syntax 1, 3, and 5 are equivalent in functionality. |

**Key Points**:
- Transposing an array reorders its dimensions according to the specified axes, without altering the size of any dimension.
- Unlike generic reshaping, transpose maintains the product of dimensions, and thus, the total number of elements remains constant.
